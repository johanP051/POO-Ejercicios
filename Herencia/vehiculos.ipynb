```python
import numpy as np
import matplotlib.pyplot as plt
import math

class Vehiculo():
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def arrancar(self):
        print("¡¡Vehículo en marcha!!")
        
        print("EL vehículo tiene un Movimiento Variado: ")
        x = np.linspace(0, 5, 100)
        y = x**2

        plt.plot(x, y)
        plt.title("Velocidad respecto al tiempo")
        plt.xlabel("t")
        plt.ylabel("x(t)")
        plt.show

        

class Moto(Vehiculo):
    def __init__(self, marca, modelo):
        super().__init__(marca, modelo)

    def hacer_caballito(self):
        confimacion = input("¿Está seguro de que desea hacer un caballito?, Puede ser peligroso. S/n")
        if confimacion.lower == "s":
            print("Haciendo un caballito")
        else:
            print("Qué bueno, puedo haber dañado la Moto si se cae")

class Camion(Vehiculo):
    def __init__(self, marca, modelo):
        super.__init__(marca, modelo)
    
    def abrir_maletero(self):
        print("El maletero se ha abierto")

carro = Vehiculo("audi", "x9")
carro.arrancar()



tipo_vehiculo = input("¿Moto o camion?")

if tipo_vehiculo.lower() == "moto":
    print("Inserte los datos de la moto: ")

    marca = input("Marca: ")
    modelo = input("Modelo: ")

    mimoto = Moto(marca, modelo)
    mimoto.arrancar()
    mimoto.hacer_caballito()
else:
    print("Inserte los datos del camión: ")

    marca = input("Marca: ")
    modelo = input("Modelo: ")

    micamion = Camion(marca, modelo)
    micamion.arrancar()
    micamion.abrir_maletero()
```
